```python
import csv
import re
name='이재명'
sentences=[]
for n in range(1,28):
    with open(f'../article/{name}/뉴스기사원문_{name}_{n}.csv','r') as f:
            file=csv.reader(f)
            next(file)
            for line in file:
                sentences.append(line[0])
```

```python
for i in range(len(sentences)):
    sentences[i]=re.sub('[-=+#/\?:^$@.,*\"※~&%ㆍ!』\\‘’|\(\◇)\[\]\<\>`\'▲▶“”…ⓒ《》【】]', ' ',sentences[i]).replace('   ','')
    sentences[i]=re.sub(r'\d{1,4}년|\d{1,2}월|\d{1,2}일', ' ',sentences[i])
    sentences[i]=re.sub(r'\d{1,4}만원|\d{1,4}억', ' ',sentences[i]).replace('   ','')
sentences_len=' '.join(sentences)
print(len(sentences_len))
print(len(sentences))
```

# 명사 추출 모델

```python
from soynlp.noun import LRNounExtractor
from soynlp.noun import LRNounExtractor_v2
from soynlp.noun import NewsNounExtractor


noun_extractor_v1 = LRNounExtractor()
nouns_v1 = noun_extractor_v1.train_extract(sentences)

noun_extractor_v2 = LRNounExtractor_v2(verbose=True)
nouns_v2 = noun_extractor_v2.train_extract(
sentences,
min_noun_score=0.3,
min_noun_frequency=20
)

noun_extractor_news = NewsNounExtractor(
    max_left_length=10, 
    max_right_length=7,
    predictor_fnames=None,
    verbose=True
)
nouns_news = noun_extractor_news.train_extract(sentences)
```

### 1. LRNounExtractor_V1 

```python
nouns_v1.keys()
```

```python
nouns_v1 = {key: value for key, value in nouns_v1.items() if len(key) > 2} 
top_v1 = sorted(nouns_v1.items(), 
    key=lambda x:-x[1].frequency * x[1].score * x[1].known_r_ratio)[:200]
for i, (word, score) in enumerate(top_v1):
    if i % 5 == 0:
        print()
    print('%6s (%.2f)' % (word, score.score), end='')
```

### 2. LRNounExtractor_V2

```python
nouns_v2.keys()
```

```python
nouns_v2 = {key: value for key, value in nouns_v2.items() if len(key) > 1}

for key in nouns_v2.keys():
    if len(key)<2:
        del nouns_v2[key]
        
top_v2 = sorted(nouns_v2.items(), 
    key=lambda x:-x[1].frequency * x[1].score)[:200]
for i, (word, score) in enumerate(top_v2):
    if i % 5 == 0:
        print()
    print('%6s (%.2f)' % (word, score.score), end='')
```

### 3. LRNounExtractor_news

```python
nouns_news.keys()
```

```python
nouns_news = {key: value for key, value in nouns_news.items() if len(key) > 2} 

tgt_noun = {}
lower_count=20
lower_score=0.3
feature_proportion=0.5

# 20 개 이상 score 0.3 이상 feature_proportion 0.5 이상 값
for noun in nouns_news:
    if (nouns_news[noun][0] >= lower_score) and (nouns_news[noun][1]) >= lower_count and (nouns_news[noun][2]) >feature_proportion:
        tgt_noun[noun]=nouns_news[noun]

top_news = sorted(tgt_noun.items(), 
    key=lambda x:-x[1].frequency * x[1].score * x[1].feature_proportion)[:200]
for i, (word, score) in enumerate(top_news):
    if i % 5 == 0:
        print()
    print('%6s (%.2f)' % (word, score.score), end='')
print(f'\n오마이뉴스 : {nouns_news["오마이뉴스"]}')
```

```python
import requests
import json
from jamo import h2j, j2hcj 
from collections import defaultdict

def get_jongsung_TF(sample_text): 
    sample_text_list = list(sample_text)
    last_word = sample_text_list[-1] 
    last_word_jamo_list = list(j2hcj(h2j(last_word))) 
    last_jamo = last_word_jamo_list[-1] 
    jongsung_TF = "T" 
    if last_jamo in ['ㅏ', 'ㅑ', 'ㅓ', 'ㅕ', 'ㅗ', 'ㅛ', 'ㅜ', 'ㅠ', 'ㅡ', 'ㅣ', 'ㅘ', 'ㅚ', 'ㅙ', 'ㅝ', 'ㅞ', 'ㅢ', 'ㅐ,ㅔ', 'ㅟ', 'ㅖ', 'ㅒ']:
        jongsung_TF = "F" 
    return jongsung_TF
```

```python
word_dic={}
lines=[]

for i in top_news:
    word_dic[i[0]]=get_jongsung_TF(i[0])
for word,TF in word_dic.items():
    line = f'{word},0,0,0,NNP,*,{TF},{word},*,*,*,*,*\n'
    lines.append(line)
with open("../vocab/vocab.csv", 'w', encoding='utf-8') as f: 
    for line in lines: 
        f.write(line)
```
